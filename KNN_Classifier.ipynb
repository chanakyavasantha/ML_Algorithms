**KNeighbor classifier Implementation**

1.**Importing required libraries**

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

Checking if there are any null values

```python
df.isnull().sum()
```

Encoding Columns

```python
le = LabelEncoder()
label = le.fit_transform(df['variety'])
print(len(label))
print(label)
```

```python
df['variety'] = label
df
```

Standardizing Data set: Subtract mean of each column from each cell and divide with S.D of each column [Column Speicific]<br>
**To Give Equal Importance to each feature**

```python
scaler = StandardScaler()
scaler.fit(df.drop('variety', axis = 1))
scaled_features = scaler.transform(df.drop('variety', axis = 1))
scaled_df = pd.DataFrame(scaled_features,columns = df.drop('variety', axis = 1).columns)
scaled_df
# Drop Target variable Column
```

```python
x = scaled_df
y = df['variety']
```

**TESTING ACCURACY FOR DIFFENRENT NUMBER OF NEIGHBORS**

```python
for i in range(1,40,2):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
    model = KNeighborsClassifier(n_neighbors = i)
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print(metrics.accuracy_score(y_test,y_pred),i,end = " ")
    print()
```
