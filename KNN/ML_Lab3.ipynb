```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
df_std = pd.DataFrame(std_scaler.fit_transform(df.drop('Target',axis = 1)),columns = df.drop('Target',axis = 1).columns)
df_std
```

```python
x = df_std
y = df['Target']
li = list(y)
print(li)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y,random_state = 42,test_size = 0.3)
X_train = np.asarray(X_train)
Y_train = np.asarray(Y_train)
X_test = np.asarray(X_test)
Y_test = np.asarray(Y_test)
```

```python
import math
from collections import Counter
import scipy.spatial
class KNN:
    def __init__(self,n):
        self.n = n
    def fit(self,x,y):
        self.X_train = x
        self.Y_train = y
    def predict(self, X_test):
        final_output = []
        for i in range(len(X_test)):
            d = []
            votes = []
            for j in range(len(X_train)):
                dist = scipy.spatial.distance.euclidean(X_train[j] , X_test[i])
                d.append([dist, j])
            d.sort()
            d = d[0:self.n]
            for d, j in d:
                votes.append(Y_train[j])
            ans = Counter(votes).most_common(1)[0][0]
            final_output.append(ans)
        return final_output            
    def Accuracy(self,yHat,y):
        cnt = 0
        print("Predicted: ")
        print(*yHat,end = " ")
        print()
        print("Actual: ")
        print(*y,end = " ")
        print()
        for i in range(len(yHat)):
            if yHat[i] == y[i]:
                cnt += 1
        print("Accuracy: ",end = " ")
        return cnt/len(yHat)
       
```

```python
for i in range(1,9,2):
    classifier = KNN(i)
    classifier.fit(X_train,Y_train)
    prediction = classifier.predict(X_test)
    print(classifier.Accuracy(prediction,Y_test))
```

**OPTMISING MODEL ACCURACY USING TECHNIQUE SIMILIAR TO BAGGING AND BOOSTING with 6 subsets**

```python
SuperSet = []
for i in range(6):
    SuperSet.append(df_std.iloc[:,i*10:i*10+10])
```

```python
Matrix = []
for SubSet in SuperSet:
    X_train,X_test,Y_train,Y_test = train_test_split(SubSet,y,random_state = 42,test_size = 0.2)
    X_train = np.asarray(X_train)
    Y_train = np.asarray(Y_train)
    X_test = np.asarray(X_test)
    Y_test = np.asarray(Y_test)
    classifier = KNN(5)
    classifier.fit(X_train,Y_train)
    prediction = classifier.predict(X_test)
    List = []
    for j in range(len(prediction)):
        List.append(prediction[j])
    Matrix.append(List)
    print(classifier.Accuracy(prediction,Y_test))
```

```python
Final = []
for li in Matrix:
    print(li,end = " ")
    print(len(li))
    print()
```

```python
Final = []
for j in range(42):
    zeroCount = 0
    oneCount = 0
    for i in range(6):
        if Matrix[i][j] == 0:
            zeroCount += 1
        else:
            oneCount += 1
    if zeroCount >= oneCount:
        Final.append(0)
    else:
        Final.append(1)
```

```python
print(*Final,end = " ")
```

```python
classifier = KNN(5)
classifier.Accuracy(Final,Y_test)
```
