```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
path = '/content/drive/MyDrive/Digit.csv'
dfTest = pd.read_csv(path)
dfTest.rename(columns = {'digitid':'Digit'},inplace = True)
dfTest
```

```python
path = '/content/drive/MyDrive/digit_train.csv'
dfTrain = pd.read_csv(path)
dfTrain.rename(columns = {'Digit-ID':'Digit'},inplace = True)
dfTrain 
```

```python
Matrix = []
for i in range(10):
    liP = []
    for j in range(7):
        oneCount = 0
        liNew = list(dfTrain['s'+str(j+1)])
        li = liNew[i*20:i*20+20]
        for k in range(len(li)):
            if li[k] == 1:
                oneCount += 1
        liP.append((1+oneCount)/22)
    Matrix.append(liP)
DF = pd.DataFrame(Matrix)
print(DF)
```

```python
MatrixFinal = []
for i in range(50):
    LI = []
    for j in range(10):
        res = 1
        for k in range(7):
            if(dfTest.iloc[i,k] == 1):
                res *= DF.iloc[j,k]
            else:
                res *= (1-DF.iloc[j,k])
        LI.append(res)
    MatrixFinal.append(LI)
DfFinal = pd.DataFrame(MatrixFinal)
DfFinal
```

```python
LIST = []
for i in range(50):
    dict = {}
    for j in range(10):
        dict[j+1] = DfFinal.iloc[i,j]
    Keymax = max(zip(dict.values(), dict.keys()))[1]
    LIST.append(Keymax)
for i in range(len(LIST)):
    if(LIST[i] == 10):
        LIST[i] = 0
print(*LIST,end = " ")
```

**ACCURACY**

```python
li = list(dfTest['Digit'])
Ans = 0
for i in range(len(li)):
    if(li[i] == LIST[i]):
        Ans += 1
print("Accuracy: ",Ans/len(li))
```
