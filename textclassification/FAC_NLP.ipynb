```python
label_encoder=preprocessing.LabelEncoder()
label_encoder.fit(df1['category'])
df1['label']=label_encoder.transform(df1['category'])
df1
```

```python
x_test = df1['body']
y_test = df1['label']
vectorized_test_x=vectorizer.fit_transform(x_test)
print(y_test)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()

X_train_cv = cv.fit_transform(x)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train_cv, y)

X_test_cv = cv.transform(x_test)
```

```python
from sklearn.metrics import accuracy_score
predictions = lr.predict(X_test_cv)
print(accuracy_score(y_test,predictions))
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train_cv, y)
X_test_cv_rf = cv.transform(x_test)
```

```python
predictions_rf = rf.predict(X_test_cv_rf)
print(accuracy_score(y_test,predictions_rf))
```
